```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('imagen1.jpeg')
b, g, r = cv2.split(img)

bhist=cv2.calcHist(b, [0], None, [256],[0, 256])
ghist=cv2.calcHist(g, [0], None, [256],[0, 256])
rhist=cv2.calcHist(r, [0], None, [256],[0, 256])
hist=cv2.calcHist([b,g,r], [0], None, [256],[0, 256])
colors=('b', 'g', 'r')

plt.figure()
plt.title("Histograma en escala RGB")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))

plt.plot(bhist)
plt.show()
plt.plot(ghist, color='g')
plt.show()
plt.plot(ghist, color='r')
plt.show()

plt.figure()
plt.title("Histograma en escala RGB")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))
for (i, col) in zip([b, g, r], colors):       # Recorre los canales de color
          hist = cv2.calcHist([i], [0], None, [256], [0, 256])   # Crea un histograma para cada canal
          plt.plot(hist, color = col)      # Grafica cada histograma
          plt.xlim([0, 256])
```
