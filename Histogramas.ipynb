```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('imagen.jpeg')
b, g, r = cv2.split(img)

bhist=cv2.calcHist([b], [0], None, [256],[0, 256])
ghist=cv2.calcHist([g], [0], None, [256],[0, 256])
rhist=cv2.calcHist([r], [0], None, [256],[0, 256])
hist=cv2.calcHist([b,g,r], [0], None, [256],[0, 256])
colors=('b', 'g', 'r')
```

```python
plt.figure()
plt.title("Histograma 'Azul'")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))

plt.plot(bhist)
plt.show()
```

```python
plt.figure()
plt.title("Histograma 'Verde'")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))

plt.plot(ghist, color='g')
plt.show()
```

```python
plt.figure()
plt.title("Histograma 'Rojo'")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))

plt.plot(rhist, color='r')
plt.show()
```

```python
plt.figure()
plt.title("Histograma en escala RGB")
plt.xlabel("Bins")
plt.ylabel("# de pixeles")
plt.xlim((0, 256))
for (i, col) in zip([b, g, r], colors):
    hist = cv2.calcHist([i], [0], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
```
