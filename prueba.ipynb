```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread("imagen1.jpeg", cv2.IMREAD_GRAYSCALE)

# Crear una matriz vacía del mismo tamaño que la imagen
imagen_estirada = np.empty_like(imagen)

# Calcular el valor mínimo y máximo de la imagen
min_val = np.min(imagen)
max_val = 200 # Limitar max_val a 200 si es mayor

# Realizar el estiramiento de histograma píxel por píxel
for i in range(imagen.shape[0]):
    for j in range(imagen.shape[1]):
        nuevo_valor = ((imagen[i, j] - min_val) / (max_val - min_val) * 255).astype(np.uint8)
        imagen_estirada[i, j] = nuevo_valor

# Mostrar la imagen original y la imagen estirada
plt.subplot(121), plt.imshow(imagen, cmap="gray"), plt.title("Original")
plt.subplot(122), plt.imshow(imagen_estirada, cmap="gray"), plt.title("Estirada")
plt.show()
```
