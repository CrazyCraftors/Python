```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('image.png')

hist_original=cv2.calcHist([imagen], [0], None, [256],[0, 256])

min = np.min(imagen)
max = 200

width, height=imagen.shape[:2]

pix = np.zeros ((width, height))

array = np.array(imagen)

for x in range(width):
    for y in range(height):
        pix [x,y] = np.mean(array[x,y])

print()

imagen_estirada=((imagen-min)/(max-min)*255)
imagen_estirada=np.uint8(imagen_estirada)

pixeles = np.zeros ((width, height))
for x in range(width):
    for y in range(height):
        pixeles[x,y]=pix[x,y]

for x in range(width):
    for y in range(height):
        imagen_estirada=((pixeles[x,y]-min)/(max-min)*255)

hist_estirado=cv2.calcHist([imagen_estirada], [0], None, [256],[0, 256])
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Histograma Original')
plt.plot(hist_original)
plt.xlim([0, 256])


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.title('Imagen Estirada')
plt.imshow(imagen_estirada)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Histograma Estirado')
plt.plot(hist_estirado)
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
