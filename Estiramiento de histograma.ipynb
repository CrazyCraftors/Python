```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('imagen.jpeg', cv2.IMREAD_GRAYSCALE)

imagen_estirada = imagen.copy()

x, y = imagen.shape

for fila in range(x):
    for columna in range(y):
        pixel = imagen[fila, columna]
        
        if pixel > 200:
            imagen_estirada[fila, columna] = 200

hist_original = cv2.calcHist([imagen], [0], None, [256], [0, 256])

min_pixel = np.min(imagen_estirada)
max_pixel = np.max(imagen_estirada)

for fila in range(x):
    for columna in range(y):
        imagen_estirada[fila, columna] = ((imagen_estirada[fila, columna] - min_pixel) / (max_pixel - min_pixel) * 255).astype(np.uint8)

hist_estirado = cv2.calcHist([imagen_estirada], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Histograma Original')
plt.plot(hist_original)
plt.xlim([0, 256])


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.title('Imagen Estirada')
plt.imshow(imagen_estirada, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Histograma Estirado')
plt.plot(hist_estirado)
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
