```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('imagen1.jpeg', cv2.IMREAD_GRAYSCALE)

hist_original = cv2.calcHist([imagen], [0], None, [256], [0, 256])

min_intensi = np.min(imagen)
max_intensi = np.max(imagen)

min_des=0
max_des=200

imagen_estirada = np.uint8((imagen-min_intensi)/(max_intensi-min_intensi)*(max_des-min_des)+min_des)

hist_estirado = cv2.calcHist([imagen_estirada], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title('Imagen Original')
plt.imshow(imagen, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Histograma Original')
plt.plot(hist_original)
plt.xlim([0, 256])

plt.subplot(2, 2, 3)
plt.title('Imagen Estirada')
plt.imshow(imagen_estirada, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Histograma Estirado')
plt.plot(hist_estirado)
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
