```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = cv2.imread('imagen1.jpeg', cv2.IMREAD_GRAYSCALE)

# Calcular el histograma de la imagen original
histograma_original = cv2.calcHist([imagen], [0], None, [256], [0, 256])

# Estirar el histograma
min_val = np.min(imagen)
max_val = np.max(imagen)
imagen_estirada = (imagen - min_val) * (255.0 / (max_val - min_val))

# Convertir la imagen de nuevo a tipo uint8
imagen_estirada = np.uint8(imagen_estirada)

# Calcular el histograma de la imagen estirada
histograma_estirado = cv2.calcHist([imagen_estirada], [0], None, [256], [0, 256])

# Mostrar la imagen original y la imagen estirada
plt.subplot(131)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')

plt.subplot(132)
plt.imshow(imagen_estirada, cmap='gray')
plt.title('Imagen Estirada')

# Mostrar los histogramas
plt.subplot(133)
plt.plot(histograma_original, color='b')
plt.plot(histograma_estirado, color='r')
plt.xlim([0, 256])
plt.legend(('Original', 'Estirado'))
plt.title('Histogramas')

plt.show()
```
