```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'imagen1.jpeg'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

shift_value = 50
shifted_image = np.copy(original_image)
shifted_image = cv2.add(shifted_image, shift_value)
shifted_image = np.clip(shifted_image, 0, 255)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(original_image.ravel(), bins=256, range=(0, 256), color='b', alpha=0.7)
plt.title('Histograma')

plt.subplot(1, 2, 1)
plt.hist(shifted_image.ravel(), bins=256, range=(0, 256), color='r', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
cv2.imshow('Imagen Original', original_image)
cv2.imshow('Imagen Desplazada', shifted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
