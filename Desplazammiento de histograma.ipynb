```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen_original = cv2.imread('imagen.jpeg', cv2.IMREAD_GRAYSCALE)

shift_value = 90
imagen_desplazada = np.copy(imagen_original)
imagen_desplazada = cv2.add(imagen_desplazada, shift_value)
imagen_desplazada = np.clip(imagen_desplazada, 0, 255)
```

```python
plt.figure(figsize=(10, 5))

plt.plot()
plt.hist(imagen_original.ravel(), bins=256, range=(0, 256), color='b', alpha=0.7)
plt.hist(imagen_desplazada.ravel(), bins=256, range=(0, 256), color='r', alpha=0.7)
plt.title('Histograma')
plt.xlabel("Bins")
plt.ylabel("# de pixeles")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(imagen_original, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(imagen_desplazada, cmap='gray')
plt.title('Imagen desplazada')
plt.axis('off')
```
